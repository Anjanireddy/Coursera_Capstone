```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
x = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup=BeautifulSoup(x, 'html.parser')
```

```python
#find table
tables=soup.find('table')
```

```python
#find rows
rows=tables.find_all('tr')
```

```python
a=[]
b=[]
c=[]
#find data in each row 
for row in rows:
    value=row.find_all('td')
    if(len(value)>0):
        a.append(value[0].text.rstrip('\n'))
        b.append(value[1].text.rstrip('\n'))
        c.append(value[2].text.rstrip('\n'))
        
```

```python
df=pd.DataFrame({"PostalCode":a,"Borough": b,"Neighborhood": c})
df.head()
```

```python
#drop cells of not assigned
df1 = df[df.Borough != "Not assigned"].reset_index(drop=True)
df1.head()
```

```python
# group neighborhoods
df2 = df1.groupby(["PostalCode", "Borough"], as_index=False).agg(lambda x: ", ".join(x))
df2.head()
```

```python
#to check the Postalcodes given in the question
x = ["M5G", "M2H", "M4B", "M1J", "M4G", "M4M", "M1R", "M9V", "M9L", "M5V", "M1B", "M5A"]

column_names = ["PostalCode", "Borough", "Neighborhood"]
test_df = pd.DataFrame(columns=column_names)

for postcode in x:
    test_df = test_df.append(df2[df2["PostalCode"]==postcode], ignore_index=True)
    
test_df
```

```python
#load the location data
location=pd.read_csv("Geospatial_Coordinates.csv")
location.head()
```

```python
# merge two tables
location.rename(columns={"Postal Code": "PostalCode"}, inplace=True)
df_new = df2.merge(location, on="PostalCode", how="left")
df_new.head()
```

```python
#to get latitude and longitude values of Toronto
address = 'Toronto'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of New York City are {}, {}.'.format(latitude, longitude))
```

## Create a map of Toronto with neighborhoods superimposed on top

```python
import folium
# create map of toronto using latitude and longitude values
map_newyork = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(df_new['Latitude'], df_new['Longitude'], df_new['Borough'], df_new['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_newyork)  
    
map_newyork
```

## let's simplify the above map and segment and cluster only the neighborhoods in Scarborough

```python
Scarborough_data=df_new[df_new['Borough'] == 'Scarborough'].reset_index(drop=True)
Scarborough_data.head()
```

```python
address = 'Scarborough'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Manhattan are {}, {}.'.format(latitude, longitude))
```

```python
# create map of Manhattan using latitude and longitude values
map_Scarborough = folium.Map(location=[latitude, longitude], zoom_start=11)

# add markers to map
for lat, lng, label in zip(Scarborough_data['Latitude'], Scarborough_data['Longitude'], Scarborough_data['PostalCode']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_Scarborough)  
    
map_Scarborough
```
